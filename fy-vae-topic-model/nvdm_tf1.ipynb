```python
import pickle

with open("nvdm/weights_1000.pkl", "rb") as f:
    weights = pickle.load(f)
```

```python
def vocab(data_url):
  """process data input."""
  data = []
  i2w = {}
  fin = open(data_url)
  while True:
    line = fin.readline()
    if not line:
      break
    word_freqs = line.split()
    index = len(data)
    data.append(word_freqs[0])
    i2w[index] = word_freqs[0]

  fin.close()
  return i2w

voc = vocab('./nvdm/data/20news/vocab.new')
```

```python
import torch

weights = torch.from_numpy(weights.T)
```

```python
weights.shape
```

```python
# check words most associated with latent dims
top_words = torch.topk(weights, 10, 0)[1]
for t in range(50):
    topic_list = []
    for w in top_words[:, t]:
        i2w = voc[w.item()]
        topic_list.append(i2w)
    print(t, topic_list)
```
