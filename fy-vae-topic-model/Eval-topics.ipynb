```python
def vocab(data_url):
  """process data input."""
  data = []
  i2w = {}
  fin = open(data_url)
  while True:
    line = fin.readline()
    if not line:
      break
    word_freqs = line.split()
    index = len(data)
    data.append(word_freqs[0])
    i2w[index] = word_freqs[0]

  fin.close()
  return i2w

voc = vocab('data/20news/vocab.new')
len(voc)
```

```python
import pickle
with open("weights_900_soft.pkl", "rb") as f:
    weights = pickle.load(f)
```

```python
# check words most associated with latent dims
import torch
for t in range(50):
    words = torch.topk(weights, 10, 0)[1][:,t]
    topic_list = []
    for w in words:
        i2w = voc[w.item()]
        topic_list.append(i2w)
    print(topic_list)
```
